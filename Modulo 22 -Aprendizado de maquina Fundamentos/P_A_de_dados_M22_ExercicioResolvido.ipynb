Neste exercício, vamos utilizar uma base de dados com informações sobre penguins. A idéia é preparar a base de dados para prever a espécie do penguin (variável resposta) baseado em suas características físicas e geográficas (variáveis preditivas).

A base de dados possui valores faltantes, utilize os conceitos da aula para trata-los.

```python
# resposta da questão 1.1

# Nos valores numericos substitui os dados faltantes pela media.
# mesmo assim há varias linha com  dados faltantes pricpalmente na coluna 
# na coluna  sex.

df = pd.DataFrame(data)

# Substituir os valores faltantes pela média em cada coluna
for coluna in df.columns:
    if df[coluna].dtype == 'float64':  # Verifica se a coluna contém números de ponto flutuante
        media_coluna = df[coluna].mean()
        df[coluna].fillna(media_coluna, inplace=True)
```

### **1.2. Variáveis numéricas** 

Identifique as variáveis numéricas e crie uma nova coluna **padronizando** seus valores. A nova coluna deve ter o mesmo nome da coluna original acrescidade de "*_std*".

> **Nota**: Você não deve tratar a variável resposta.

```python
# resposta da questão 1.2

# Variavel resposta sera "species" - A especie do pinguin !

# colunas normalizada Dp nessa etapa 
# bill_length_mm , bill_depth_mm , flipper_length_mm , body_mass_g !.

# coletando o desvio padrão das variaveis numericas 
df_limp.head(n=1)

md_length = df_limp['bill_length_mm'].mean()
md_depth = df_limp['bill_depth_mm'].mean()
md_flipper = df_limp['flipper_length_mm'].mean()
md_body = df_limp['body_mass_g'].mean()

dp_length = df_limp['bill_length_mm'].std()
dp_depth = df_limp['bill_depth_mm'].std()
dp_flipper = df_limp['flipper_length_mm'].std()
dp_body = df_limp['body_mass_g'].std()
```

```python
# Retornando as novas colunas padronizadas
# aplicando a  media e desvio padrõa 
df_limp['bill_length_mm_std'] = df_limp['bill_length_mm'].apply(lambda mm : (mm - md_length) / dp_length)
df_limp['bill_depth_mm_std'] = df_limp['bill_depth_mm'].apply(lambda mm : (mm - md_depth) / dp_depth)
df_limp['flipper_length_mm_std'] = df_limp['flipper_length_mm'].apply(lambda mm : (mm - md_flipper) / dp_flipper)
df_limp['body_mass_g_std'] = df_limp['body_mass_g'].apply(lambda mm : (mm - md_body) / dp_body)
```

```python
#df_limp.head(n=10)

teste = df_limp[['flipper_length_mm_std']].max()
print(teste)
```

### **1.3. Variáveis categóricas** 

Identifique as variáveis categóricas nominais e ordinais, crie uma nova coluna aplicando a técnica correta de conversão a seus valores. A nova coluna deve ter o mesmo nome da coluna original acrescidade de "*_nom*" ou "*_ord*".

> **Nota**: Você não deve tratar a variável resposta.

Variável respota "species"

```python
# resposta da questão 1.3

# Variaveis Categoricas norminais e ordnais 
# São as de tipo string 


# Atributos categóricos ordinais são aqueles em que os valores apresentem relação de ordem.
# Visto que as variavel island não apresenta ordem o codigo a baixo seria mesmo o ideal ?
# visto que uma ilha não e maior ou melhor que a outra 

#df_limp['island'].drop_duplicates()

island_mapper = {
    "Torgersen": 0,
    "Biscoe": 1,
    "Dream" : 2
}


df_limp['island_ord'] = df_limp['island'].apply(lambda level : island_mapper[level])
```

```python
# respota mais aceitavel para a coluna "island"
# nesse atividade não apresenta categoria ordinal

islanda_teste = pd.get_dummies(df_limp['island'])
islanda_teste.columns = [ str(col) + '_nrm'  for col in islanda_teste.columns]
print(islanda_teste)



df_limp = pd.concat([df_limp,islanda_teste], axis=1)

# Quase esqueci da coluna "sex"

sex_m = pd.get_dummies(df_limp['sex'])
sex_m.columns = [ str(col) + '_nrm'  for col in sex_m.columns]
print(sex_m)



df_limp = pd.concat([df_limp,sex_m], axis=1)
```
